```python
import cv2
import numpy as np
import time

import requests

def download_file(url, filename):
    """
    Hàm để tải file từ url và lưu vào filename
    """
    r = requests.get(url, allow_redirects=True)
    open(filename, 'wb').write(r.content)

# Các URL chứa file mà bạn muốn tải về
weights_url = 'https://pjreddie.com/media/files/yolov3.weights'
config_url = 'https://raw.githubusercontent.com/pjreddie/darknet/master/cfg/yolov3.cfg'

# Các tên file mà bạn muốn lưu trên máy
weightsPath = 'yolov3.weights'
configPath = 'yolov3.cfg'

# Tải file về
download_file(weights_url, weightsPath)
download_file(config_url, configPath)

# Load the YOLO network
net = cv2.dnn.readNetFromDarknet(configPath, weightsPath)

layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers().flatten()]

# Initialize the video stream and allow the camera sensor to warmup
cap = cv2.VideoCapture(0)

count = 0
save_path = r"C:\Users\dat20\Downloads"

# Initialize the time
start_time = time.time()

# Define colors for different classes
colors = np.random.uniform(0, 255, size=(len(layer_names), 3))

while True:
    # Capture frame-by-frame
    _, frame = cap.read()

    # Prepare the frame to run through the deep learning model
    height, width, channels = frame.shape
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)

    # pass the blob through the network and obtain the detections
    net.setInput(blob)
    outs = net.forward(output_layers)

    # loop over the detections
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                
                # Rectangle coordinates
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                
                # Draw the rectangle and label on the frame
                color = colors[class_id]
                cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
                cv2.putText(frame, str(class_id), (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

                # Save the captured image into the given directory every 6 seconds (10 frames per minute)
                if time.time() - start_time >= 6:
                    cv2.imwrite(f"{save_path}/frame_{count}.jpg", frame)
                    count += 1
                    start_time = time.time()

    # Display the resulting frame
    cv2.imshow('Video', frame)

    # if the `q` key is pressed, break from the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```
