```python
# Cài đặt các thư viện cần thiết
!pip install opencv-python-headless scikit-image matplotlib

import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt

from google.colab import files

def compare_images(imageA, imageB, title):
    # Tính toán độ tương tự (SSIM) giữa hai hình ảnh
    score = ssim(imageA, imageB, multichannel=True)

    # Hiển thị hình ảnh và độ tương tự
    fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

    ax[0].imshow(cv2.cvtColor(imageA, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Image A')

    ax[1].imshow(cv2.cvtColor(imageB, cv2.COLOR_BGR2RGB))
    ax[1].set_title('Image B')

    # Hình ảnh chồng lên nhau
    overlay = cv2.addWeighted(imageA, 0.5, imageB, 0.5, 0)
    ax[2].imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
    ax[2].set_title('Overlay: SSIM: %.2f' % score)

    # Hiển thị kết quả
    plt.show()

# Tải lên hai hình ảnh từ máy tính
uploaded = files.upload()

# Lưu tên file hình ảnh
image_files = list(uploaded.keys())

# Đọc hình ảnh
imageA = cv2.imread(image_files[0])
imageB = cv2.imread(image_files[1]) if len(image_files) > 1 else imageA

# Đảm bảo cả hai hình ảnh đều có cùng kích thước bằng cách điều chỉnh kích thước của hình ảnh thứ hai để phù hợp với hình ảnh thứ nhất
imageB = cv2.resize(imageB, (imageA.shape[1], imageA.shape[0]))

# So sánh hai hình ảnh
compare_images(imageA, imageB, "Image A vs. Image B")


# So sánh hai hình ảnh
compare_images(imageA, imageB, "Image A vs. Image B")
```
